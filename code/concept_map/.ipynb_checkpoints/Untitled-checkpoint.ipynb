```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE

algos = pd.read_csv("rl_algos.csv")
algos.head()

concepts = algos['name']
features = algos.drop('name', axis=1)

tsne = TSNE(n_components=3, perplexity=1)
embedded_features = tsne.fit_transform(features)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(embedded_features[:, 0], embedded_features[:, 1], embedded_features[:, 2], c='r', marker='o')

for i, concept in enumerate(concepts):
    ax.text(embedded_features[i, 0], embedded_features[i, 1], embedded_features[i, 2], concept, fontsize=8)

plt.show()
```
